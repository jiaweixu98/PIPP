```python
# load the twitter data from json files 
pddict = {'time':[],'fulltext':[]}
from tqdm import tqdm
import os
import json
import gzip
from dateutil import parser
for filename in tqdm(os.listdir('2020')):
    if 'Tweet' in filename:
        with gzip.open('2020/%s'%filename, mode="rt") as f:
            for line in f:
                tweet = json.loads(line)
                pddict['time'].append(parser.parse(tweet['created_at']))
                pddict['fulltext'].append(tweet['full_text'])
```

```python
import pandas as pd
df = pd.DataFrame(data=pddict)
df
```

```python
df.to_csv('out.csv')
df
```
