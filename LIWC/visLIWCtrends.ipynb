```python
import pandas as pd
aus_comm = pd.read_csv('LIWC-22 Results - out - LIWC Analysis.csv')
```

```python
aus_comm['time_1'] = pd.to_datetime(aus_comm['time_1'])
```

```python
aus_comm.dtypes
```

```python
aus_comm = aus_comm.set_index('time_1')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(11, 4)})
```

```python
aus_comm['Year'] = aus_comm.index.year
aus_comm['Month'] = aus_comm.index.month
```

```python
data_columns = ['Analytic', 'Clout', 'Authentic', 'Tone','emo_anx','emo_anger','emo_sad']
aus_comm_weekly_mean = aus_comm[data_columns].resample('W').mean()
```

```python

# Start and end of the date range to extract
start, end = '2020-01', '2021-01'
# Plot daily and weekly resampled time series together
fig, ax = plt.subplots()
ax.plot(aus_comm_weekly_mean.loc[start:end, 'Analytic'],
marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.set_ylabel('Analytic Thinking')
ax.legend()
ax.set_title('Analytic Thinking in austin')
```
